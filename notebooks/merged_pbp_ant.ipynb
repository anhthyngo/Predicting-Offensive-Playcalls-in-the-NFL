Merging the play-by-play data for regular season + post season from 2009 - 2019

Data from: https://github.com/ryurko/nflscrapR-data/tree/master/play_by_play_data


```python
import os
import pandas as pd
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path = '/Users/anhthyngo/ds-projects/nfl-predict/data/play_by_play_data'
dr = os.listdir("/Users/anhthyngo/ds-projects/nfl-predict/data/play_by_play_data")

```

```python
dfs = []

for file in dr:
    if file.endswith('.csv'):
        print(file)
        dfs.append(pd.read_csv(os.path.join(path, file)))
```

```python
merged = pd.concat(dfs, axis=0)
merged.shape
```

```python
data_path = '/Users/anhthyngo/ds-projects/nfl-predict/data/'
```

```python
merged.to_csv(data_path+'2009-2019-play-by-play.csv')
```
