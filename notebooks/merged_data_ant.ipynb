new_pbp and pbp dataframe can be joined by game_id and GameId respectively.  
This way we can retrieve the home_team and create a new unique key that is a composite of game_id + home_team to merge with the weather dataset ant scraped.

```python
weather.head()
```

```python
## map each team to url name
team_mapping = {
    'PHI': 'phi',
    'BAL': 'rav',
    'NYG': 'nyg',
    'NO' : 'nor',
    'NE' : 'nwe',
    'MIN': 'min',
    'MIA': 'mia',
    'IND': 'clt',
    'CLE': 'cle',
    'LAC': 'sdg',
    'DEN': 'den',
    'CAR': 'car',
    'ARI': 'crd',
    'GB' : 'gnb',
    'DET': 'det',
    'OAK': 'rai',
    'CIN': 'cin',
    'ATL': 'atl',
    'WAS': 'was',
    'TEN': 'oti',
    'TB' : 'tam',
    'PIT': 'pit',
    'NYJ': 'nyj',
    'BUF': 'buf',
    'SF' : 'sfo',
    'LA' : 'ram',
    'JAX': 'jax',
    'DAL': 'dal',
    'CHI': 'chi',
    'HOU': 'htx',
    'KC' : 'kan',
    'SEA': 'sea'
    
}
```

```python
## use the mapping above to create new unique id for game_id which is a composite of gamedate + home team
new_pbp = new_pbp.replace({"home_team": team_mapping})
new_pbp['game_id'] = new_pbp['game_id'].astype(str).str[0:9] + new_pbp['home_team']
new_pbp.head()
        
        
```

```python
merged_data = new_pbp.merge(weather, left_on = 'game_id', right_on = 'game_id')
```

```python
merged_data.head(2)
```

```python
merged_data['year'] = merged_data['game_id'].astype(str).str[0:4]
```

```python
## let's only take data since 2014.. i don't think we need that much anyways
mask = merged_data[merged_data['year'].astype('int32') >= 2014]
mask.head(1)
```
